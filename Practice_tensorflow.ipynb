```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
from keras.utils import plot_model
print(tf.__version__)
```

```python
tf.random.set_seed(42) # Setting a global seed
a1 = tf.random.uniform([3,2], minval=4, maxval=10)
a2 = tf.random.uniform([3,2], minval=4, maxval=10)
print(a1)
print(a2)
tf.random.set_seed(42) # Reseting the global seed
b1 = tf.random.uniform([3,2], minval=4, maxval=10)
b2 = tf.random.uniform([3,2], minval=4, maxval=10)
print(b1)
print(b2)
# so here a1 = b1 and a2 = b2

####INTUITION####
# Think of tf.random.set_seed() like rewinding a playlist:
# First play ‚Üí song1 = a1, song2 = a2.
# Reset ‚Üí playlist starts over ‚Üí song1 = b1, song2 = b2.
# That‚Äôs why the pairs match.
```

```python
# Now for Generator class
gen1 = tf.random.Generator.from_seed(42) # Intuition think of 42 as a playlist of song
gen2 = tf.random.Generator.from_seed(42) # 42 is a playlist of song
z1 = gen1.uniform([3,2], minval=4, maxval=10) # it gives first song of the playlist
z2 = gen1.uniform([3,2], minval = 4, maxval = 10) # it gives second song of the playlist
z3 = gen2.uniform([3,2], minval=4, maxval=10) # it also gives first song of the same playlist
print(z1)
print(z2)
print(z3)
```

```python
##DECORATOR###
def decorator(func):
  def wrapper():
    print("Start")
    func()
    print("End")
  return wrapper

@decorator
def hello():
  print("Hello World")
# So the wrapper delays the execution of the decorated function until it‚Äôs actually called.
# Without the wrapper, your function runs at decoration time, not at call time.hello()
```

```python
import time

def add_tf_uniform(x, y):
    return x + y

@tf.function
def add_tf_deco_uniform(x, y):
    return x + y

x = tf.random.uniform([1000, 1000])
y = tf.random.uniform([1000, 1000])

# Without tf.function
start = time.time()
for _ in range(100):
    add_tf_uniform(x, y)
end = time.time()
print("Without tf.function decorator time =", end - start)

# With tf.function
start = time.time()
for _ in range(100):
    add_tf_deco_uniform(x, y)
end = time.time()
print("With tf.function decorator time =", end - start)

# add_tf_uniform: runs eagerly (Python executes line by line).
# add_tf_deco_uniform: runs as a TensorFlow graph (compiled once, reused).
# First call of add_tf_deco_uniform might be a little slower (graph tracing).
# But repeated calls are faster compared to eager execution.
```

# Using the various methods of tf.keras.metrics.MeanAbsoluteError()
In Multiple Batches

```python
mae = tf.keras.metrics.MeanAbsoluteError()
# Batch 1
y_true_1 = [1,2,3]
y_pred_1 = [1.1, 2.2, 3.3]
mae.update_state(y_true_1, y_pred_1) ####
print(mae.result()) ####
print(mae.result().numpy()) ####
```

```python
# Batch 2
y_true_2 = [1,2,3]
y_pred_2 = [1.2, 2.2, 3.3]
mae.update_state(y_true_2, y_pred_2)
print(mae.result())
print(mae.result().numpy())
```

```python
mae.reset_state() ####
print(mae.result())
print(mae.result().numpy())
```

# MEAN SQUARED ERROR

# Checking the difference between tf.keras.metrics.MeanSquaredError()
# and tf.keras.losses.MeanSquaredError()

```python
# Batch 1 using metrics sub-modlule
y_true_1 = tf.constant([1,2,3], dtype = tf.float32)
y_pred_1 = tf.constant([1.1, 2.2, 3.3])

mse_metric = tf.keras.metrics.MeanSquaredError()
mse_metric.update_state(y_true_1, y_pred_1)
# print(mse_metric.result())
print(f"From the metrics sub-module MSE: {mse_metric.result().numpy()}")

# Now for the same Batch 1 we are using losses sub-module
mse_loss = tf.keras.losses.MeanSquaredError()
loss = mse_loss(y_true_1, y_pred_1)
print(f"From the losses sub-module MSE: {loss.numpy()}")
```

```python
# Now for the Second Batch of data
# Batch 2 using metrics sub-module
y_true_2 = tf.constant([1,2,3], dtype = tf.float32)
y_pred_2 = tf.constant([1.2, 2.2, 3.3])

mse_metric.update_state(y_true_2, y_pred_2)
print(f"From the metrics sub-module MSE: {mse_metric.result().numpy()}")

# Now using losses sub-module for the same Batch 2
loss2 = mse_loss(y_true_2, y_pred_2)
print(f"From the losses sub-module MSE: {loss2.numpy()}")

# """ FROM THE RESULT WE CAN SEE THAT LOSSES SUB-MODULE IS NOT STORING INTERNAL STATE AND IT IS WORKING ON CURRENT BATCH
#     WHILE, METRICS SUB-MODULE IS SAVING THE INTERNAL STATE AND GIVES ACCUMULATED RESULT. """
```
